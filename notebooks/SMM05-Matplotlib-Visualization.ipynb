# Evaluating Model Performance with Visualizations:
* Use KNN model on Iris data, find optimal value of k 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=9876)
```

### Generate values of k - set of number 1-99, and take odd values
* Use arange() function, start point, end point, counting by every other number
* Train classifier on each of these values of k
* Use accuracy method for test set, and save to list

```python
k_values = np.arange(1, 100, 2)
print(k_values)
```

```python
accuracies = []

for k in k_values:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_true = y_test
    accuracy = accuracy_score(y_true, y_pred)
    accuracies.append(accuracy)
```

## Plot Model Accuracy Metrics as a Function of K-values

```python
plt.plot(k_values, accuracies)
plt.title("Model Accuracy as a Function of K-values")
plt.xlabel("Values of K")
plt.ylabel("Accuracy")
plt.show()
```

```python
k_accuracies = list(zip(k_values, accuracies))
k_accuracies[5:15]
```

### Optimal values of k for KNN model are between 17 to 21 neighbors
